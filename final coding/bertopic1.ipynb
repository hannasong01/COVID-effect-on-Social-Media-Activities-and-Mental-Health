```python
dp_pre_txt = dp[["Title"]]
dp_post_txt = dp_post[["Title"]]
anx_pre_txt = anx_pre[["Title"]]
anx_post_txt = anx_post[["Title"]]
adhd_pre_txt = adhd_pre[["Title"]]
adhd_post_txt = adhd_post[["Title"]]
bi_pre_txt = bi_pre[["Title"]]
bi_post_txt = bi_post[["Title"]]
pt_pre_txt = pt_pre[["Title"]]
pt_post_txt = pt_post[["Title"]]
sc_pre_txt = sc_pre[["Title"]]
sc_post_txt = sc_post[["Title"]]
ed_pre_txt = ed_pre[["Title"]]
ed_post_txt = ed_post[["Title"]]
bf_pre_txt = bf_pre[["Title"]]
bf_post_txt = bf_post[["Title"]]
sw_pre_txt = sw_pre[["Title"]]
sw_post_txt = sw_post[["Title"]]
```

```python
pre_txt = pd.concat([dp_pre_txt, anx_pre_txt, adhd_pre_txt, bi_pre_txt, pt_pre_txt, sc_pre_txt, ed_pre_txt, bf_pre_txt, sw_pre_txt])
post_txt = pd.concat([dp_post_txt, anx_post_txt, adhd_post_txt, bi_post_txt, pt_post_txt, sc_post_txt, ed_post_txt, bf_post_txt, sw_post_txt])
```

```python
pre_txt['Post Text'].fillna('', inplace=True)
pre_txt.dropna(subset=['Post Text'], inplace=True)
post_txt['Post Text'].fillna('', inplace=True)
post_txt.dropna(subset=['Post Text'], inplace=True)

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

pre_txt['Post Text'] = pre_txt['Post Text'].apply(remove_punctuations)
post_txt['Post Text'] = post_txt['Post Text'].apply(remove_punctuations)

pre_txt['Post Text'] = pre_txt['Post Text'].replace('\n','', regex=True)
post_txt['Post Text'] = post_txt['Post Text'].replace('\n','', regex=True)

pre_txt['Post Text'] = pre_txt['Post Text'].str.lower()
post_txt['Post Text'] = post_txt['Post Text'].str.lower()
```

```python
pre_txt['Title'].fillna('', inplace=True)
pre_txt.dropna(subset=['Title'], inplace=True)
post_txt['Title'].fillna('', inplace=True)
post_txt.dropna(subset=['Title'], inplace=True)

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

pre_txt['Title'] = pre_txt['Title'].apply(remove_punctuations)
post_txt['Title'] = post_txt['Title'].apply(remove_punctuations)

pre_txt['Title'] = pre_txt['Title'].replace('\n','', regex=True)
post_txt['Title'] = post_txt['Title'].replace('\n','', regex=True)

pre_txt['Title'] = pre_txt['Title'].str.lower()
post_txt['Title'] = post_txt['Title'].str.lower()
```
